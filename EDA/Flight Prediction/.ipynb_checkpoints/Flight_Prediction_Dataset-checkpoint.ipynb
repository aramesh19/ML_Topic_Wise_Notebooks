```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df = pd.read_excel('Data_Train.xlsx')
train_df.head()
```

```python
test_df = pd.read_excel('Test_set.xlsx')
test_df.head()
```

```python
final_df = train_df.append(test_df)
final_df.tail()
```

```python
final_df.info()
```

```python
## Checking null values
final_df.isnull().sum()
```

```python
final_df['Date_of_Journey'].str.split('/').str[0]
```

```python
##### Feature engineering process
final_df['Date'] = final_df['Date_of_Journey'].apply(lambda x:x.split('/')[0])
final_df['Month'] = final_df['Date_of_Journey'].apply(lambda x:x.split('/')[1])
final_df['Year'] = final_df['Date_of_Journey'].apply(lambda x:x.split('/')[2])
```

```python
final_df.head()
```

```python
final_df['Date'] = final_df['Date'].astype(int)
final_df['Month'] = final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
final_df['Arrival_Time'] = final_df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
final_df.head(5)
```

```python
final_df['Arrival_hour'] = final_df['Arrival_Time'].apply(lambda x:x.split(':')[0])
final_df['Arrival_min'] = final_df['Arrival_Time'].apply(lambda x:x.split(':')[1]) 
```

```python
final_df.head(1)
```

```python
final_df['Arrival_hour'] = final_df['Arrival_hour'].astype(int)
final_df['Arrival_min'] = final_df['Arrival_min'].astype(int)
```

```python
final_df.info()
```

```python
final_df.head(1)
final_df.drop('Arrival_Time',axis=1,inplace=True)
final_df.head(1)
```

```python
final_df['Dep_hour'] = final_df['Dep_Time'].apply(lambda x:x.split(':')[0])
final_df['Dep_min'] = final_df['Dep_Time'].apply(lambda x:x.split(':')[1])
```

```python
final_df.drop('Dep_Time',axis=1,inplace=True)
final_df.head(1)
```

```python
final_df['Dep_hour'] = final_df['Dep_hour'].astype(int) 
final_df['Dep_min'] = final_df['Dep_min'].astype(int)
```

```python
final_df.info()
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops'] = final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```

```python
final_df.drop('Route',axis=1,inplace=True)
```

```python
final_df['Duration_hour'] = final_df['Duration'].apply(lambda x:x.split('h')[0])
```

```python
final_df['Duration_hour']
```

```python
final_df[final_df['Duration_hour']=='5m']
```

```python
final_df.drop(final_df[final_df['Duration_hour']=='5m'].index,inplace=True)
```

```python
final_df['Duration_hour'] = final_df['Duration_hour'].astype(int)
```

```python
final_df.info()
```

```python
final_df['Duration_min'] = final_df['Duration'].apply(lambda x:x.split('h')[1])
```

```python
final_df['Duration_min'] = final_df['Duration_min'].apply(lambda x:x.split('m')[0]) 
```

```python
final_df['Duration_min'].astype(int)
```

```python
final_df[final_df['Duration_min'] == '']
```

```python
final_df.loc[final_df['Duration_min']=='','Duration_min']=0
```

```python
final_df[final_df['Duration_min'] == 0]
```

```python
final_df['Duration_min'] = final_df['Duration_min'].astype(int)
```

```python
final_df.head()
```

```python
final_df.drop('Duration',axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
final_df['Airline'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
final_df['Airline'] = le.fit_transform(final_df['Airline'])
final_df['Source'] = le.fit_transform(final_df['Source'])
final_df['Destination'] = le.fit_transform(final_df['Destination'])
final_df['Additional_Info'] = le.fit_transform(final_df['Additional_Info'])
```
