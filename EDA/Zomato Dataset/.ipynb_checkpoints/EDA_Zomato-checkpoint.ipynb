```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df_final = pd.merge(df,df_country,on='Country Code', how='left')
```

```python
df_final.head()
```

```python
#DataTypes
df_final.dtypes
```

```python
df_final.columns
```

```python
country_names=df_final.Country.value_counts().index
```

```python
country_val = df_final.Country.value_counts().values
```

```python
## Pie chart
plt.pie(country_val,labels=country_names)
```

```python
plt.pie(country_val[:3],labels=country_names[:3])
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.1f%%')
```

```python
df_final.columns
```

```python
ratings = df_final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (11,5)
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings, hue='Rating color', palette=['white','red','orange','yellow','lightgreen','darkgreen'])
```

```python
## Count plot
sns.countplot(x='Rating color', data=ratings, palette=['white','red','orange','yellow','lightgreen','darkgreen'])
```

```python
## Find the countries that have given zero rating
zerorating = df_final.groupby(['Aggregate rating','Country']).size().reset_index()
zerorating
```

```python
df_final[df_final['Rating color'] == 'White'].groupby('Country').size().reset_index()
```

```python
## find out which currency is used by which country
df_final.groupby(['Country','Currency']).size().reset_index()
```

```python
## which countries have online deliveries option
df_final.groupby(['Has Online delivery','Country']).size().reset_index()
```

```python
##  Create a piechart for cities distribution
city_values = df_final.City.value_counts().values
city_labels = df_final.City.value_counts().index
plt.pie(city_values[:5],labels=city_labels[:5], autopct='%1.1f%%')
```

```python
# Find the top 10 quisines 
df_final.Cuisines.value_counts()
```
